# Media Tweet Analysis
1. There is some clutering of the compound value of the sentiments for all the media sources at 1 with some spread of data points as depicted in the scatter plot. 
2. BBC and CBS have a positive average compound score for the anlysis and indicate a positive sentiment(bar chart) 
3. Fox news, CNN  and NYtimes have a negative average compound score indicating a negative sentiment for their last 100 tweets(bar chart). 

```python
# Dependencies
import tweepy
import json
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
from config import (consumer_key, 
                    consumer_secret, 
                    access_token, 
                    access_token_secret)

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())

# Output File (CSV)
output_data_file = "output_data/news_tweets.csv"
```

```python
# Target User Accounts
target_users = ("@BBC", "@CBS", "@CNN", "@FoxNews", "@nytimes")
# List for dictionaries of results
sentiments = []
# Counter
counter = 1
oldest_tweet = None


# Loop through each user
for user in target_users:

    # Variables for holding sentiments
    compound_list = []
    positive_list = []
    negative_list = []
    neutral_list = []
    
  # Loop through 5 pages of tweets (total 100 tweets)
    for x in range(5):

        # Get all tweets from home feed
        public_tweets = api.user_timeline(user, page=x)

        # Loop through all tweets
        for tweet in public_tweets:
            # Print Tweets
            # print("Tweet {counter}: {tweet["text"]})

            # Run Vader Analysis on each tweet
            results = analyzer.polarity_scores(tweet["text"])
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]
            
             # Get Tweet ID, subtract 1, and assign to oldest_tweet
            oldest_tweet = tweet['id'] - 1

            # Add each value to the appropriate list
            compound_list.append(compound)
            positive_list.append(pos)
            negative_list.append(neg)
            neutral_list.append(neu)
            
            # Append to sentiments list

            sentiments.append({ "Username": user,
                           "Date": tweet["created_at"], 
                           "Compound": compound,
                           "Positive": pos,
                           "Negative": neu,
                           "Neutral": neg,
                           "Tweets Ago": counter})
            # Add to counter 
            counter += 1
```

```python
# Create DataFrame from Results List
sentiments_pd = pd.DataFrame(sentiments)
sentiments_pd.head()
```

```python
# drop the @in username 
sentiments_pd["Username"] = sentiments_pd["Username"].map(lambda x: str(x)[1:])
sentiments_pd.head()
```

```python
#Save as CSV
sentiments_pd.to_csv(output_data_file)
```

```python
#Setting the data frames by the Usernames, converting the time stamp strings and 
#sorting by date and time for the latest 100 tweets 

#BBC
bbc = sentiments_pd.loc[sentiments_pd["Username"] == "BBC", :]
bbc = pd.DataFrame(bbc)

#Convert tweet timestamps to datetime objects that can be manipulated by
# Python
   
bbc['Date'] = pd.to_datetime(bbc.Date)
 
bbc['Date']= bbc['Date'].dt.strftime('%a %b %d %H:%M:%S %z %Y')

# sort by time stamps
bbc_sorted = bbc.sort_values('Date', ascending = False)
bbc_sorted = pd.DataFrame(bbc_sorted)

bbc_sorted = bbc_sorted.assign(tweet=[1 + i for i in range(len(bbc_sorted))])[['tweet'] + bbc_sorted.columns.tolist()]

#Set x, y coordinates for sorted data
bbc_tweetsago = bbc_sorted["tweet"]
bbc_compound = bbc_sorted["Compound"]



#CBS
cbs = sentiments_pd.loc[sentiments_pd["Username"] == "CBS", :]
cbs = pd.DataFrame(cbs)

#Convert tweet timestamps to datetime objects that can be manipulated by
# Python
   
cbs['Date'] = pd.to_datetime(cbs.Date)
 
cbs['Date']= cbs['Date'].dt.strftime('%a %b %d %H:%M:%S %z %Y')

# sort by time stamps
cbs_sorted = cbs.sort_values('Date', ascending = False)
cbs_sorted = pd.DataFrame(cbs_sorted)
cbs_sorted = cbs_sorted.assign(tweet=[1 + i for i in range(len(cbs_sorted))])[['tweet'] + cbs_sorted.columns.tolist()]

#Set x, y coordinates for sorted data
cbs_tweetsago = cbs_sorted["tweet"]
cbs_compound = cbs_sorted["Compound"]


#CNN
cnn = sentiments_pd.loc[sentiments_pd["Username"] == "CNN", :]
cnn = pd.DataFrame(cnn)

#Convert tweet timestamps to datetime objects that can be manipulated by
# Python
   
cnn['Date'] = pd.to_datetime(cnn.Date)
 
cnn['Date']= cnn['Date'].dt.strftime('%a %b %d %H:%M:%S %z %Y')

# sort by time stamps
cnn_sorted = cnn.sort_values('Date', ascending = False)
cnn_sorted = pd.DataFrame(cnn_sorted)
cnn_sorted = cnn_sorted.assign(tweet=[1 + i for i in range(len(cnn_sorted))])[['tweet'] + cnn_sorted.columns.tolist()]

#Set x, y coordinates for sorted data
cnn_tweetsago = cnn_sorted["tweet"]
cnn_compound = cnn_sorted["Compound"]

#fox
fox = sentiments_pd.loc[sentiments_pd["Username"] == "FoxNews", :]
fox = pd.DataFrame(fox)

#Convert tweet timestamps to datetime objects that can be manipulated by
# Python
   
fox['Date'] = pd.to_datetime(fox.Date)
 
fox['Date']= fox['Date'].dt.strftime('%a %b %d %H:%M:%S %z %Y')

# sort by time stamps
fox_sorted = fox.sort_values('Date', ascending = False)
fox_sorted = pd.DataFrame(fox_sorted)

fox_sorted = fox_sorted.assign(tweet=[1 + i for i in range(len(fox_sorted))])[['tweet'] + fox_sorted.columns.tolist()]

#Set x, y coordinates for sorted data
fox_tweetsago = fox_sorted["tweet"]
fox_compound = fox_sorted["Compound"] 

#NY Times
ny = sentiments_pd.loc[sentiments_pd["Username"] == "nytimes", :]
ny = pd.DataFrame(ny)

#Convert tweet timestamps to datetime objects that can be manipulated by
# Python
   
ny['Date'] = pd.to_datetime(ny.Date)
 
ny['Date']= ny['Date'].dt.strftime('%a %b %d %H:%M:%S %z %Y')

# sort by time stamps
ny_sorted = ny.sort_values('Date', ascending = False)
ny_sorted = pd.DataFrame(ny_sorted)

ny_sorted = ny_sorted.assign(tweet=[1 + i for i in range(len(ny_sorted))])[['tweet'] + ny_sorted.columns.tolist()]


#Set x, y coordinates for sorted data

ny_tweetsago = ny_sorted["tweet"]
ny_compound = ny_sorted["Compound"]   

ny_sorted.head()
```

```python
# Build the scatter plots for each news group
bbc_scatter = plt.scatter(bbc_tweetsago,
            bbc_compound, 
            c="skyblue", 
            s=100,
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="BBC")

cbs_scatter = plt.scatter(cbs_tweetsago, 
            cbs_compound, 
            c="green", 
            s=100,
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="CBS")

cnn_scatter = plt.scatter(cnn_tweetsago, 
            cnn_compound, 
            c="red", 
            s=100,
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="CNN")

fox_scatter = plt.scatter(fox_tweetsago, 
            fox_compound, 
            c="blue", 
            s=100,
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Fox News")

nyt_scatter = plt.scatter(ny_tweetsago, 
            ny_compound, 
            c="Yellow",
            s=100,
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="New York Times")

plt.legend(handles=[bbc_scatter, cbs_scatter, cnn_scatter, fox_scatter, nyt_scatter], title='Media Sources',
           loc='upper center', bbox_to_anchor=(0.5,-0.1))

x_limit = 105
now = datetime.now()
now = now.strftime("%Y-%m-%d %H:%M")
plt.title(f"Sentiment Analysis of Media Tweets ({now})")
plt.ylabel("Tweet Polarity")
plt.xlabel("Tweets Ago")
plt.xlim(x_limit,0)
plt.show()


# Save Figure
plt.savefig("output_data/sentiments_scatter.png")


# Show plot
plt.show()
```

```python
#BAR CHART
# summarize data for bar chart
user_group =sentiments_pd.groupby("Username")
compound_average = user_group["Compound"].mean().round(3)
compound_average 
```

```python
#Overall Media Sentiment Bar Chart
colors = ['skyblue','green','red', 'blue', 'yellow']
sentiment_chart = compound_average.plot(kind='bar', color=colors, align ='center',width=1.0, edgecolor ='b')

# Set the xlabel and ylabel 
sentiment_chart.set_xlabel("Media Sources")
sentiment_chart.set_ylabel("Tweet Polarity")

#Bar value labels 
ax=sentiment_chart
#annotate axis = seaborn axis
for p in ax.patches:
             ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11, color='black', xytext=(0, 20),
                 textcoords='offset points')
        
sentiment_chart.set_ylim(-0.1010,0.25) #To make space for the annotations 

#Other properties of the chart
now = datetime.now()
now = now.strftime("%Y-%m-%d %H:%M")
plt.title(f"Overall Media Sentiment based on twitter ({now})") 
plt.ylabel("Tweet Polarity")
plt.xlabel("Media Sources")
plt.grid(False)
plt.show()
plt.savefig("output_data/sentiments_bar.png")
 
```
